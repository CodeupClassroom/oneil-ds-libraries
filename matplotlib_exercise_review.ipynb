#### Exercise 1

Use matplotlib to plot the following equation:
    
$y = x^2 - x + 2$

```python
# Generate data with range 

x = range(-100, 100)
y = [(num**2 - num + 2) for num in x]

```

```python
# Plot x and y with plot()
plt.plot(x,y)

# Add title
plt.title('$y = x^2 - x +2$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$y$')

# Annotate (0,0) with coordinate
plt.annotate('Origin', xy = (0,0), xytext = (0, 3000), arrowprops ={})

# Set y limit
plt.ylim(0,10500)

# save the figure
# Uncomment if desired
#plt.savefig('figure1.png')

plt.show()
```

#### Exercise 2

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
    
$y = \sqrt(x)$

$y = x^3$

$y = 2^x$
    
$y = \dfrac{1}{x+1}$

```python
# Generate x1 and y1 values

x1 = range(0,100)
y1 = [math.sqrt(n) for n in x1]

# plot the data
plt.plot(x1,y1)

# Customize
plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--');

plt.show()
```

```python
# Generate x2 and y2 values
x2 = range(0,20)
y2 = [n**3 for n in x2]

#plot data
plt.plot(x2,y2)

# Customize
plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');
```

```python
# Generate x3 and y3 values
x3 = [x * 0.1 for x in range(0,150)]
y3 = [2 ** n for n in x3]

# plot data
plt.plot(x3,y3)

# Customize
plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.grid(True, ls = '--')

# Add ticks as well
plt.xticks(range(0, 16));

```

```python
# Note 0.1x is used for a smoother curve
# x4 and y4 
x4 = [x * 0.1 for x in range(0,150)]
y4 = [1/(n+1) for n in x4]


# Plot the data
plt.plot(x4,y4)


# Customize
# Add title
plt.title('$ y = \dfrac{1}{x+1}$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$\dfrac{1}{x+1}$')


plt.grid(True, ls = '--')

plt.show()
```

Combine the four figures from exercises 2 into one large figure using subplots.

### Pie Chart

- Shows proportion of data
- Uses frequency 


```python
titanic = data('titanic')
```

```python
titanic.sample(5)
```

```python
survivors = sum(titanic.survived =='yes')
non_survivors = sum(titanic.survived =='no')
```

```python
print(f'{survivors} survivors \n\n{non_survivors} non-survivors ')
```

```python
# Assign labels
my_labels = ['Survived', 'Did not Survive']

# Set the size of each part
sizes = [survivors, non_survivors]

# Create the figure
# Create the subplot (ax)

fig, ax = plt.subplots()
ax.pie(sizes, labels = my_labels, autopct = '%1.0f%%', colors = ('tan', 'firebrick'))

ax.set_title('Survival Rate for Titanic');

```
