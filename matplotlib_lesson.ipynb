### Chart Types

### Line Charts

- Line charts connect values with a line
- Shows changes in values over time
- Time values should not have gaps


### Scatter plots

- Scatter plots display values as separate points
- Used for two continuous numerical variables
- Markers represent the paired values $(x, y)$



### Histograms

[Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

- Shows distribution of data
- Values are sorted into **bins**
- $x$-axis is divided into intervals based on number of bins
- $y$-axis plots the count (frequency) of values within the bin

### Example 1
    
    
The data plotted below measures the bacterial growth for E. coli. This bacteria can double in number every 20 minutes. 
    
- Create titles and axis labels for the plot below.
- Customize the chart however you see fit.
    

````python
# Create data

# Set time values in increments of 20 
time = range(1, 600, 20)

## Set initial number of bacteria equal to 1
initial = 1


bacteria = [ initial * np.float_power(2, (0.05*n)) for n in time ]
````

### Example
    
Create two plots to represent the sine and cosine function. Add a legend to describe the two plots. 

````python
x = range(-4, 5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14, 6)) # (width, height)

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', label='$\cos(x)$')

plt.legend(loc=4)
plt.title('Approximate sin and cosine curves')
plt.xlabel('$x$')


plt.show()
````

---
### Default histogram example using `plt.hist()`

````python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.hist(x)
plt.show()
````

## Bins and alignment

- Bin values are the intervals in which values are placed. 
- The width of the bin is the difference between the start and end of a bin
- Bin alignment set how the bars are aligned relative to the bin values


````python
## Data
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

# bin width is 1
# number of bins is 6

plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6])
plt.grid(False)
plt.show()
````

<div class="alert alert-block alert-warning">

### Example - Bin alignment

- Create three histograms as subplots.
- Set the alignment value to 'default', 'left', and 'right'

````python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]

plt.figure(figsize=(16, 6))
plt.suptitle(f'x = {x}')

plt.subplot(132)
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6])
plt.title('align="middle" (default)')

plt.subplot(131)
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6], align='left')
plt.title('align="left"')

plt.subplot(133)
plt.hist(x, bins=[0, 1, 2, 3, 4, 5, 6], align='right')
plt.title('align="right"')

plt.show()
````

### Example - Overlapping histograms

Use the transparency value `alpha = ` to show both histograms.

````python
x1 = [random.randint(1, 5) for _ in range(20)]
x2 = [random.randint(1, 5) for _ in range(20)]

plt.hist(x1, bins=[0, 1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.5, color='blue')
plt.hist(x2, bins=[0, 1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.5, color='blue')

plt.show()
````

<div class="alert alert-block alert-warning">

### Example - Side-by-side histograms

- Assign two different sets of values to `.hist()`
    - use a tuple (data1, data2)
- Create a legend
- Set a color for the edges with `edgecolor = `

````python
# Note that _ is an index for the loop
x1 = [random.randint(1, 5) for _ in range(20)]
x2 = [random.randint(1, 5) for _ in range(20)]

# Use (x2, x2)
plt.hist((x1, x2), bins=[1, 2, 3, 4, 5, 6], align='left', edgecolor='black', alpha=0.6, label=['$x_1$', '$x_2$'])

plt.legend()
plt.show()
````

### Example - Subplots with histograms

- Create 4 histograms
- Each will represent the sum of two dice which are rolled N times
    - N = 10
    - N = 100
    - N = 1000
    - N = 10000
- Generate roll values randomly
- Add subplot titles
- Add a figure title

````python
plt.figure(figsize=(12, 8))
for i in range(1, 5):
    n = 10 ** i
    x1 = [random.randint(1, 6) for _ in range(n)]
    x2 = [random.randint(1, 6) for _ in range(n)]

    total = [x + y for x, y in zip(x1, x2)]

    plt.subplot(2, 2, i)
    plt.hist(total, bins=range(2, 14), align='left', color='white', edgecolor='black')
    plt.title(f'n = {n}')

plt.suptitle('Sum of 2 dice rolls')
plt.show()
````
